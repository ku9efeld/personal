# Выгрузка по всем городам, поиск по топ-20 городов
## Создание временной таблицы

```python
cursor.execute('DROP TABLE IF EXISTS #city_cord')
#
sql = '''
-- Создаем временную таблицу
CREATE TABLE #city_cord(
city VARCHAR(100),
box_coordinate GEOMETRY
)
'''
#
cursor.execute(sql)
```

## Города и координаты

```python
cities = ['Paris',
          'Baghdad',
          'Havana',
          'Istanbul',
          'Tbilisi',
          'Zurich',
          'Berlin',
          'Tirana',
          'Vienna',
          'Milan',
          'Rome',
          'Vilnius',
          'Amsterdam',
          'Warsaw',
          'Casablanca',
          'Barcelona',
          'Antwerp',
          'Hamburg',
          'Belgrade',
          'Lyon']
data = [ox.geocode_to_gdf(city).total_bounds for city in  cities]
dict_cities = dict(zip(cities,data))
insert_array = []
for city in dict_cities.keys():
    insert_array.append( (city, box(*dict_cities[city]).wkt) )
```

# Вставка во временную таблицу

```python
insert_array
```

```python
%%time
sql = '''INSERT INTO #city_cord VALUES (%s, ST_GeomFromText(%s))'''
cursor.executemany(sql,insert_array)
```

# Топ городов

```python
%%time
sql = '''
WITH points AS (SELECT
        ST_MakePoint
        (
        CAST(params.lon AS double precision),
        CAST(params.lat AS  double precision)
        ) AS point,
        params.category
FROM TABLE
WHERE event_at::date BETWEEN '2023-09-01' AND '2023-09-01' 
AND event_name = 'objectAddFavourites'
)

SELECT
    COUNT (*) as total_by_category,
    city,
    category
FROM(SELECT 
    point,
    (SELECT city FROM #city_cord WHERE ST_CONTAINS(box_coordinate,point) IS TRUE LIMIT 1) AS city,
    category
FROM points)
GROUP BY city,category
ORDER BY total_by_category DESC
'''
cursor.execute(sql)
top_bookmark = cursor.fetch_dataframe()
```

```python
top_bookmark.to_csv('top_20_city_bookmarks')
```
