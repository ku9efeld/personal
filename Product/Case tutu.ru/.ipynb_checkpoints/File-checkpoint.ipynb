1. Существуют значительная доля людей, совершивших покупку без рассмотра баннера, поэтому конверсия уходит > 100
 
Нужно установить какая доля людей от всех покупок совершила покупку без клика баннера и по клику баннера

```python
profiles = df\
           .query('title == "order"')\
           .sort_values(by='time')\
           .groupby('user')\
           .agg(first_purchase_dt=('time','first'),
                product=('product','first')
               )\
           .reset_index()
```

```python
profiles
```

```python
profiles['product'].value_counts().plot(kind='bar');
```

```python
events = df[df['user'].isin(profiles['user'])]
```

```python
clicks = events.query('title =="banner_click"').groupby('user').size().rename('total_clicks',inplace=True)
```

```python
profiles = pd.merge(profiles,clicks,how='left',on='user')
```

```python
(profiles['total_clicks'].isna().sum()*100/profiles.shape[0]).round(2)
```

Порядка 53% пользователей, которые совершили покупку - не кликали по баннеру до покупки.  

Теперь нужно установить, сколько пользователей, из которых кликали 1-2 раза (значительная доля пользователей) - их клик не ассоциируется с реальной их покупкой.
 + Из идей - можно добавить рядом количество покупок

```python
events[['next_title','next_title_dt']] = events.sort_values(by=['user','time']).groupby('user')[['title','time']].shift(-1)
```

```python
events['dif_dt'] = (events['next_title_dt'] - events['time']).dt.total_seconds()/(60*60)
```

```python
sns.boxplot(data=events.query('title=="banner_click" and next_title=="order"'),x='dif_dt')
```

```python
fig,axes = plt.subplots(ncols=5,figsize=(10,6),sharey=True)
data=events.query('title=="banner_click" and next_title=="order"')\
    .groupby('product')\
    .boxplot(column='dif_dt',
             ax=axes)
for ax in axes:
    ax.axhline(y=1,linestyle='--')
```
