#  Расчет выборок, длительности и предварительная оценка мощности и значимости

```python
from scipy.stats import norm
# Вычисления N
def calc_n(baseline,mde,alpha,power):
    beta = 1-power
    z_a=norm.ppf(1-alpha/2)
    z_b=norm.ppf(1-beta)
    var1=baseline*(1-baseline)
    var2=(baseline+mde)*(1-baseline-mde)
    return np.ceil((z_a*np.sqrt(2*var1)+z_b*np.sqrt(var2+var1))**2/(mde*mde))
# Вверхняя оценка размер выборок
def calc_n_approx(alpha,power,mde):
    beta = 1-power
    z_a=norm.ppf(1-alpha/2)
    z_b=norm.ppf(1-beta)
    return np.ceil(((z_a+z_b)/mde))**2
# Таблица
def aprox_table(powers,betas,alpha=0.05):
    resultTable = []
    for beta,power in zip(betas,powers):
        n = []
        for mde in mdes:
            n.append(calc_n_approx(0.05,power,mde))
        resultTable.append(n)
    resultDF = pd.DataFrame(resultTable,index=powers,columns=mdes).rename_axis(index='power',columns='MDE')
    return resultDF
def aprox_table_(baseline,powers,betas,alpha=0.05):
    resultTable = []
        for beta,power in zip(betas,powers):
            n = []
            for mde in mdes:
                n.append(calc_n(baseline,mde,0.05,power))
    resultTable.append(n)
    resultDF = pd.DataFrame(resultTable,index=powers,columns=mdes).rename_axis(index='power',columns='MDE')
    return resultDF
```

```python
from statsmodels.stats.proportion import proportions_ztest, proportion_confint,confint_proportions_2indep
import scipy.stats as stats
# Сводная таблица по ОС
def stats_matrix(os_frame):
    conversion_rates = os_frame.groupby('group')['converted']
    conversion_rates = conversion_rates.agg([np.mean, np.sum, np.size])
    conversion_rates.columns = ['conversion_rate', 'successes','n']
    return conversion_rates
# Попарное сравнение выборок
def test_matrix(control,test,alpha=0.05,side='two-sided'):
    successes = [control.successes,test.successes]
    nobs = [control.n, test.n]
    if side == 'two-sided':
        z_stat, pval = proportions_ztest(successes, nobs=nobs)
        low,up = confint_proportions_2indep(count1=successes[0],count2=successes[1],nobs1=nobs[0],nobs2=nobs[1],alpha=alpha)
        print(control.conversion_rate - test.conversion_rate)
        print(f'z statistic: {z_stat:.6f}')
        print(f'p-value: {pval:.3f}')
          print(f'ci: {low:.3f},{up:.3f}')
    if side == 'one':
        z_stat, pval = proportions_ztest(successes, nobs=nobs,alternative='smaller')
        z_crit = stats.norm.cdf(1-alpha)
        print('Smaller' if control.conversion_rate < test.conversion_rate else 'Larger')
        print(f'z statistic: {z_stat:.2f}')
        print(f'p-value: {pval:.3f}')
        print(f'ci interval: ({z_crit:.3f}, inf+)')
```
