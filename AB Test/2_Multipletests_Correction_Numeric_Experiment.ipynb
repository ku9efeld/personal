# Множественное тестирование



**Reference:**

```python
np.random.seed(5)
alpha = 0.05 # FWER
sample_size = 100
l_groups = 18
g_groups = 5
#
norm_dist_1 = [sps.norm.rvs(0,1,sample_size) for _ in range(l_groups)]
#
norm_dist_2 = [sps.norm.rvs(0.4,1,sample_size) for _ in range(g_groups)]
#
group_label_d1 = [['l_'+str(i)]*sample_size for i in range(l_groups)]
#
group_label_d2 = [['g_'+str(i)]*sample_size for i in range(g_groups)]
#
samples = np.append(np.array(norm_dist_1).ravel(),np.array(norm_dist_2).ravel())
#
labels = np.concatenate(group_label_d1 + group_label_d2)
#
comp = MultiComparison(samples, labels)
```

```python
fig = plt.figure(figsize=(12,9))
plt.title(r"$X^l\;и\;\widetilde{X}^g$",fontsize=16)
pd.DataFrame(zip(samples,labels),columns=['value','group']).groupby('group').boxplot(subplots=False,boxprops=dict(linestyle='-', linewidth=1.5))
labels_x = [t_c.get_text()[1:5].replace(',','').replace('_','=') for t_c in plt.gca().get_xticklabels()]
plt.ylabel('X')
plt.xlabel('Индекс выборки')
plt.gca().set_xticklabels(labels_x)
None
```

```python
# Выполнение поправки Бонферрони
bonferroni = comp.allpairtest(sps.ttest_ind,
                              alpha=alpha,
                              method='bonf')
# Выполнение поправки Шидака
sidak = comp.allpairtest(sps.ttest_ind,
                         alpha=alpha,
                         method='sidak')
# Выполнение поправки Холма
holm = comp.allpairtest(sps.ttest_ind,
                        alpha=alpha,
                        method='holm')
```

```python

class MultiTestCrossTab():
    p_value_ticks = [0.01,0.05,0.25,0.5,0.75,1]
    
    def __init__(self,result):
        self.df = pd.DataFrame(result)
        self.df['test_case'] = self.df.apply(lambda x: 'M(X)=M(Y)' if (x['group1'].startswith('g') and x['group2'].startswith('g')) or (x['group1'].startswith('l') and x['group2'].startswith('l')) else 'M(X)~=M(Y)',axis=1)
        self.df['reject_no_cor'] = self.df.apply(lambda x: True if x['pval'] < 0.05 else False,axis=1)
      # Две таблицы
        self.crosstab_cor  = pd.crosstab(self.df['reject'],self.df['test_case'])
        self.crosstab_default = pd.crosstab(self.df['reject_no_cor'],self.df['test_case'])
    def view(self,mode='cor'): # mode - cor - "correction"; default -"no correction"
        if mode == 'cor':
            return self.crosstab_cor
        elif mode == 'def':
            return self.crosstab_default
        elif mode == 'source':
            return self.df
```

```python
MultiTestCrossTab(bonferroni[2]).view('def')
#/(math.comb(l_groups+g_groups, 2)
```

```python
import matplotlib.pyplot as plt
import math
# Default
sort_pval = MultiTestCrossTab(bonferroni[2]).view('source').sort_values(by='pval').reset_index(drop=True)
sort_pval['color'] =  sort_pval['reject_no_cor'].apply(lambda x: 'r' if x == True else 'b')
# Другое
scatter=plt.scatter(x=sort_pval.index,y=sort_pval.pval,c=sort_pval.color,alpha=0.25,s=6,marker='o')
#
plt.yticks([0.01,0.05,0.25,0.5,0.75,1])
plt.grid(True,linestyle=':')
plt.xlabel('i')
plt.title('Отсортированные достигаемые уровни значимости (n={})'.format(math.comb(l_groups+g_groups, 2)))
plt.axhline(y=alpha,linewidth=0.2,linestyle='-.',color='black')
plt.ylim(ymax=1.02)
plt.ylabel('p-value')
None
```

```python
data = MultiTestCrossTab(bonferroni[2]).view('source').sort_values(by='pval_corr').reset_index(drop=True)
data['color']=((data['test_case'] == 'M(X)=M(Y)')).apply(lambda x: 'b' if x == True else 'r')
plt.scatter(x=data.index,y=data.pval_corr,c=data.color,alpha=0.25,s=10,marker='o')




# Визуализация
plt.title('Поправка Бонферрони')
plt.yticks([0.01,0.05,0.25,0.5,0.75,1])
plt.ylim(ymax=1.02)
plt.ylabel(r'$\widetilde{p}($i$)$')
plt.xlabel(r'$i$')
plt.grid(True,linestyle=':')
```

```python
_,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
axes=axes.flatten()
m = math.comb(l_groups+g_groups, 2)
#
data = MultiTestCrossTab(holm[2]).view('source').sort_values(by='pval_corr').reset_index(drop=True)
data['color']=((data['test_case'] == 'M(X)=M(Y)')).apply(lambda x: 'b' if x == True else 'r')
axes[0].scatter(x=data.index,y=data.pval_corr,c=sort_pval.color,alpha=0.25,s=10,marker='o')


plt.scatter([i for i in range(m)],[(alpha/(m-i)) for i in range(m)],s=8)



# Визуализация
axes[0].set_ylabel(r'$\widetilde{p}($i$)$')
axes[0].set_title('Поправка Холма')
axes[0].set_yticks([0.01,0.05,0.25,0.5,0.75,1])
axes[0].set_ylim(ymax=1.02)
axes[0].set_xlabel(r'$i$')
axes[0].grid(True,linestyle=':')
#
axes[1].set_ylabel(r'$\alpha_i$')
axes[1].set_xlabel(r'$i$')
axes[1].set_title('Уровни значимости с поправкой')
axes[1].grid(True,linestyle=':')
```

```python
_,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
axes=axes.flatten()

data = MultiTestCrossTab(sidak[2]).view('source').sort_values(by='pval_corr').reset_index(drop=True)
data['color']=((data['test_case'] == 'M(X)=M(Y)')).apply(lambda x: 'b' if x == True else 'r')
axes[0].scatter(x=data.index,y=data.pval_corr,c=sort_pval.color,alpha=0.25,s=10,marker='o')

axes[1].scatter([i for i in range(m)],[1-np.power(1-alpha,1/(m-i)) for i in range(m)],s=8)


# Визуализация
axes[0].set_ylabel(r'$\widetilde{p}($i$)$')
axes[0].set_title('Поправка Шидака')
axes[0].set_yticks([0.01,0.05,0.25,0.5,0.75,1])
axes[0].set_ylim(ymax=1.02)
axes[0].set_xlabel(r'$i$')
axes[0].grid(True,linestyle=':')
#
axes[1].set_ylabel(r'$\alpha_i$')
axes[1].set_xlabel(r'$i$')
axes[1].set_title('Уровни значимости с поправкой')
axes[1].grid(True,linestyle=':')
```
